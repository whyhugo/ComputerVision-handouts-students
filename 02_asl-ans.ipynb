```python
import pandas as pd

train_df = pd.read_csv("asl/sign_mnist_train.csv")
valid_df = pd.read_csv("asl/sign_mnist_valid.csv")
```

### æ¢ç´¢è³‡æ–™

```python
train_df.head()
```

### æ“·å–å½±åƒé¡åˆ¥

```python
y_train = train_df['label']
y_valid = valid_df['label']
del train_df['label']
del valid_df['label']
```

### æ“·å–å½±åƒ

```python
x_train = train_df.values
x_valid = valid_df.values
```

## è¦–è¦ºåŒ–è³‡æ–™

æŠŠè³‡æ–™(image) 1D to 2D

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(40,40))

num_images = 20
for i in range(num_images):
    row = x_train[i]
    label = y_train[i]
    
    image = row.reshape(28,28)
    plt.subplot(1, num_images, i+1)
    plt.title(label, fontdict={'fontsize': 30})
    plt.axis('off')
    plt.imshow(image, cmap='gray')
```

## ç·´ç¿’ï¼šæ­£è¦åŒ–å½±åƒè³‡æ–™

```python
x_train.min()
```

```python
x_train.max()
```

```python
# TODO: Normalize x_train and x_valid.
x_train = x_train / 255
x_valid = x_valid / 255
```

## ç·´ç¿’ï¼šå½±åƒåˆ†é¡

Hint: ä½¿ç”¨ [keras.utils.to_categorical](https://www.tensorflow.org/api_docs/python/tf/keras/utils/to_categorical) æ–¹æ³•

```python
import tensorflow.keras as keras
num_classes = 24
```

```python
# TODO: Categorically encode y_train and y_valid.
y_train = keras.utils.to_categorical(y_train, num_classes)
y_valid = keras.utils.to_categorical(y_valid, num_classes)
```

## ç·´ç¿’ï¼šå»ºç«‹æ¨¡å‹

We need...
* å…·æœ‰å¯†é›†è¼¸å…¥å±¤ã€‚é€™ä¸€å±¤åŒ…å« 512 å€‹ç¥ç¶“å…ƒï¼Œä½¿ç”¨ `relu`æ¿€æ´»å‡½æ•¸ï¼Œä¸¦æ¥æ”¶å½¢ç‹€ç‚º `(784,)` çš„è¼¸å…¥å½±åƒã€‚
* å…·æœ‰ç¬¬äºŒå€‹å¯†é›†è¼¸å…¥å±¤ï¼Œå«æœ‰ 512 å€‹ä½¿ç”¨ `relu`æ¿€æ´»å‡½æ•¸çš„ç¥ç¶“å…ƒ
* å…·æœ‰å¯†é›†è¼¸å‡ºå±¤ï¼Œå…¶ç¥ç¶“å…ƒçš„æ•¸é‡èˆ‡é¡åˆ¥æ•¸é‡ç›¸ç­‰ï¼Œä¸¦ä½¿ç”¨ `softmax`æ¿€æ´»å‡½æ•¸

é‚„ä¸ç†Ÿæ‚‰ï¼Œåƒè€ƒä¸€ä¸‹mnistçš„ç¯„ä¾‹å§~

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# TODO: build a model following the guidelines above.
model = Sequential()
model.add(Dense(units = 512, activation='relu', input_shape=(784,)))
model.add(Dense(units = 512, activation='relu'))
model.add(Dense(units = num_classes, activation='softmax'))
```

## ç¸½çµæ¨¡å‹

```python
model.summary()
```

## ç·¨å¯«æ¨¡å‹

ä½¿ç”¨[CategoricalCrossentropy](https://www.tensorflow.org/api_docs/python/tf/keras/losses/CategoricalCrossentropy)

```python
model.compile(loss='categorical_crossentropy', metrics=['accuracy'])
```

## ç·´ç¿’ï¼šè¨“ç·´æ¨¡å‹

ä½¿ç”¨æ¨¡å‹çš„ `fit`æ–¹æ³•ä¾†å°æ¨¡å‹é€²è¡Œ 20 å€‹ Epoch çš„è¨“ç·´ï¼Œä½¿ç”¨çš„æ˜¯å…ˆå‰å»ºç«‹çš„è¨“ç·´èˆ‡é©—è­‰å½±åƒå’Œé¡åˆ¥ï¼š

```python
# TODO: Train the model for 20 epochs.
model.fit(x_train, y_train, epochs=20, verbose=1, validation_data=(x_valid, y_valid))
```
